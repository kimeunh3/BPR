```python
import pandas as pd
import numpy as np
```

```python
names = ['user_id', 'item_id', 'rating', 'timestamp']
data = pd.read_csv("./data/u.data", delimiter='\t', names=names)
data = data.drop(["rating", "timestamp"], axis=1)
data
```

```python
items = data["item_id"].unique()
items
```

```python
data = data.groupby("user_id")["item_id"].apply(list).reset_index(name='positive')
```

```python
n_sample = 2
data["negative"] = data["positive"].apply(lambda x: list(set(items) - set(x)))
data
```

```python
data = data.explode("positive")
data
```

```python
data["negative"] = data["negative"].apply(lambda x: np.random.choice(x, n_sample))
data
```

```python
data.explode("negative")
```

```python
data["n_pos"] = data["positive"].apply(len)
data["n_neg"] = data["negative"].apply(len)
data["total"] = data["n_pos"] + data["n_neg"]
data
```
